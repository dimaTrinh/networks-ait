## Related notations and formulas:

- $S$ = relative size of the largest component = $ \frac{s_{1}}{N} \approx 1 - e^{1-<k>S}$

## Code

```python
import networkx as nx;
import numpy as np;
import matplotlib;
import matplotlib.pyplot as plt;
import math;
```

```python
N = 1000
p = 1/(N-1)
nw_before_crit = nx.generators.erdos_renyi_graph(N,p-0.001)
nw_after_crit = nx.generators.erdos_renyi_graph(N,p+0.001)
```

```python
plt.figure(figsize = (10, 8))
nx.draw(nw_before_crit, node_size = 20)
plt.show()
```
