## Topic 

Study the percolation transition of the E-R graph numerically. How
does the size of the largest connected component scale with the system
size at the critical point? How does the size of the largest connected
component grow with $p$ at the critical point?
In the presentation show pictures of the generated graphs both below
and above the critical point, and also figures revealing the scaling of the largest connected component at $p_{c}$.

## Related notations and formulas:
- $<k>$ = average degree in network $ = (N-1)p$.
- $S$ = relative size of the largest component = $ \frac{s_{1}}{N} \approx 1 - e^{1-<k>S}$

## Code

```python
import networkx as nx;
import numpy as np;
import matplotlib;
import matplotlib.pyplot as plt;
import math;
```

```python
def Calc_chi_and_S(network):
    _network_N = network.number_of_nodes();
    _comps = nx.connected_components(network);
    _comp_sizes = [len(_comp) for _comp in _comps];
    _sort_c_sizes = sorted(_comp_sizes,reverse=True);
    _lcs = _sort_c_sizes[0]/_network_N;
    _chi = 0;
    if len(_sort_c_sizes) > 1:
        _chi = sum([_sort_c_sizes[i]*_sort_c_sizes[i] for i in range(1,len(_sort_c_sizes))]);
        return _chi/(len(_sort_c_sizes)),_lcs;
    else:
        return _chi,_lcs;
```

```python
N, num_samp = 1000, 20;
delta = 0.001
est_p_crit = 0;
maxChi = 0;
p_list = np.linspace(1/(N-1)-delta, 1/(N-1)+delta, 300);
av_S,av_chi = [],[];
av_S.clear();
av_chi.clear();
for p in p_list:
    S_values, chi_values = [],[];
    S_values.clear();
    chi_values.clear();
    for i in range(0,num_samp):
        ER_graph = nx.generators.erdos_renyi_graph(N,p);
        chi,S = Calc_chi_and_S(ER_graph);
        S_values.append(S);
        chi_values.append(chi);
    av_S.append(np.mean(S_values));
    meanChi = np.mean(chi_values)
    if (meanChi > maxChi):
        maxChi = meanChi;
        est_p_crit = p;
    av_chi.append(meanChi);
```

```python
import json

data = {}
data['N'] = N;
data['est_p_crit'] = est_p_crit;
data['maxChi'] = maxChi;
data['p_list'] = p_list.tolist();
data['av_S'] = av_S;
data['av_chi'] = av_chi;

with open('s_chi_e-r_1000_scale_p.json', 'w') as json_file:
    json.dump(data, json_file)
```

```python
#setting the interval
p_nei = p_list[begInt:endInt+1]
s_nei = av_S[begInt:endInt+1]
```

```python
def getRMSE(fitted, x, y):
    return np.mean((y- fitted(x)) ** 2)
```

```python
est_S = {} #storing the estimated values by each model
err_model = [] #storing the error for each model
degRange = range(2,6)
p_iter = np.linspace(p_nei[0], p_nei[len(p_nei)-1], 1000)
for deg in degRange:
  #fitting the linear model
  model = np.poly1d(np.polyfit(p_nei, s_nei, deg))

  #getting the estimated results
  est_S[deg] = model(p_iter)

  #getting the RMSE for each model
  err_model.append(getRMSE(model, p_nei, s_nei))
```

```python
mult = 1.0
plt.clf()
plt.figure(figsize = (12, 10))
plt.xlabel(xlabel = r'$p$', fontsize = 14)
plt.ylabel(ylabel = r'$S=\frac{s_1}{N}$', fontsize = 14)
plt.plot(p_nei,s_nei,'-',label = 'S')
plt.xlim((p_nei[0], p_nei[len(p_nei)-1]))
for deg in degRange:
    plt.plot(p_iter, est_S[deg], label = "Linear model with degree " + str(deg))
plt.legend(loc = 0)
plt.grid()
plt.show()
```

```python
mult = 1.0
plt.clf()
plt.xlabel(xlabel = r'Degree', fontsize = 14)
plt.ylabel(ylabel = r'RMSE', fontsize = 14)
plt.plot(list(degRange), err_model)
plt.grid()
plt.show()
```

```python
from scipy.optimize import curve_fit
#Source: https://towardsdatascience.com/basic-curve-fitting-of-scientific-data-with-python-9592244a2509
```

```python
# Function to calculate the power-law with constants a and b
def power_law(x, a, b):
    return a*np.power(x, b)
```

```python
pars, cov = curve_fit(f=power_law, xdata=p_nei, ydata=s_nei, p0=[0, 0], bounds=(-np.inf, np.inf), maxfev = 1000)
pow_model_est = power_law(p_iter, *pars)
```

```python
plt.clf()
fig,ax = plt.subplots()

# Set the x and y-axis scaling to logarithmic
ax.set_xscale('log')
ax.set_yscale('log')
plt.xlabel(xlabel = r'$\log(p)$', fontsize = 14)
plt.ylabel(ylabel = r'$\log(S)$', fontsize = 14)

# Set the axis limits
plt.plot(p_nei, s_nei, label = 'Actual S')
plt.plot(p_iter, pow_model_est, label = 'Power law model')
plt.legend(loc = 0)
plt.show()
```
